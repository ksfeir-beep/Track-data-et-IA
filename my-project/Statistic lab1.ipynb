Iris flowers

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
feature_names = iris.feature_names
df = pd.DataFrame(iris.data, columns=feature_names)

# Map species
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = pd.Series(iris.target).map(species_map)

# Reorder columns
df = df[['species'] + feature_names]

# Summary statistics for all flowers
summary_all = df.describe(include='all')
print("Descriptive statistics for all flowers:\n")
print(summary_all)

# Summary statistics by species
summary_by_species = df.groupby('species').describe()
print("\nDescriptive statistics for each species:\n")
print(summary_by_species)


import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")

# Scatter plot: Sepal length vs Sepal width
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='Set1', s=100)
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()

# Scatter plot: Petal length vs Petal width
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', hue='species', palette='Set1', s=100)
plt.title("Petal Length vs Petal Width by Species")
plt.show()

# Liste des colonnes de caractéristiques
features = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

# Créer un boxplot pour chaque caractéristique
for feature in features:
    plt.figure(figsize=(8,6))
    sns.boxplot(x='species', y=feature, data=df, palette='Set2')
    plt.title(f'Boxplot de {feature} par espèce')
    plt.show()
```
